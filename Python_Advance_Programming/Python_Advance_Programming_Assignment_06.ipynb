1. You are given two strings s and t. String t is generated by randomly shuffling string s and then adding one more letter at a random position. Return the letter that was added to t.

```python
def find_the_difference(in_one,in_two):
    in_two_temp = list(in_two)
    for ele in in_one:
        if ele in in_two_temp:
            in_two_temp.pop(in_two_temp.index(ele))
    return ','.join(in_two_temp)
```

```python
find_the_difference("abcd", "abcde") 
```

```python
find_the_difference("", "y")
```

```python
find_the_difference("ae", "aea") 
```

2. Given a function that accepts unlimited arguments, check and count how many data types are in those arguments. Finally return the total in a list.

```python
from collections import OrderedDict
def count_datatypes(*args):
    
    output = OrderedDict({'int':0,'str':0,'bool':0,'list':0,'tuple':0,'dict':0})
    for ele in args:
        output[type(ele).__name__] += 1
        output_list = list(output.values())
    return list(output.values())
        
```

```python
count_datatypes(1, 45, "Hi", False)  
```

```python
A Fibonacci string is a precedence of the Fibonacci series. It works with any two characters of the English alphabet (as opposed to the numbers 0 and 1 in the Fibonacci series) as the initial items and concatenates them together as it progresses in a similar fashion as the Fibonacci series.
```

```python
def fib_str_one(in_num,in_list):
    out_list = in_list.copy()
    if in_num>2:
        for ele in range(2,in_num):
            out_list.append(out_list[-1]+out_list[-2])
    return ", ".join(out_list)
```

```python
fib_str_one(3, ["j", "h"]) 
```

```python
fib_str_one(5, ["e", "a"])
```

```python
fib_str_one(6, ["n", "k"])
```

4. Given an integer between 0 and 26, make a variable (self.answer). That variable would be assigned to a string in this format:

```python
def ones_threes_nines(in_num):
    in_num_clone = in_num
    answer = {'nines':0,'threes':0,'ones':0}
    if in_num>0 and in_num<26:
        while in_num !=0:
            if in_num>=9:
                in_num -= 9
                answer['nines'] += 1
            elif in_num>=3:
                in_num -= 3
                answer['threes'] +=1
                
            elif in_num >= 1:
                in_num -= 1
                answer['ones'] += 1
                
        print(f"ones_threes_nines({in_num_clone}) ➞ \"nines:{answer['nines']}, threes:{answer['threes']}, ones:{answer['ones']}\"")
        return answer
```

```python
ones_threes_nines(10)
```

```python
ones_threes_nines(15)
```

```python
ones_threes_nines(22)
```
